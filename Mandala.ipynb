```python
import numpy as np
import matplotlib.pyplot as plt

def Mandala(multiplier, rad):
    t = np.arange(0, 2 * np.pi, 0.01)
    x0 = np.sin(multiplier * t)
    y0 = np.cos(multiplier * t)
    x1 = rad * np.sin(13 * t)
    y1 = rad * np.cos(13 * t)

    lines = np.column_stack((x0, x1, y0, y1)).reshape(-1, 2)
    plt.figure(figsize=(10,10)); plt.plot(*lines, color='black')
    plt.gca().set_aspect('equal', adjustable='box');  plt.axis('off')
    plt.show()
```

```python
#examples:

Mandala(100, 455565)
Mandala (8, 0.72)


```
